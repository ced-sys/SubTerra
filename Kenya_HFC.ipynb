```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
df=pd.read_excel('/content/IHFC_2024_GHFDB.xlsx', sheet_name='data release2024')
print(df.columns)
```

```python
df['P1']=pd.to_numeric(df['P1'], errors='coerce')
df['P5']=pd.to_numeric(df['P5'], errors='coerce')
df['P4']=pd.to_numeric(df['P4'], errors='coerce')

df.dropna(subset=['P1', 'P5', 'P4'])
```

```python
gdf=gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df['P4'], df['P5']),
    crs='EPSG:4326'
)
```

```python
import os
os.environ['SHAPE_RESTORE_SHX']='YES'
kenya=gpd.read_file('/content/gadm41_KEN_0.shp')
```

```python
gdf_kenya=gdf[gdf.geometry.within(kenya.unary_union)]
```

```python
gdf_kenya.to_file('heat_flow_kenya.geojson', driver='GeoJSON')
gdf_kenya[['P1', 'P4', 'P5']].to_csv('heat_flow_kenya.csv', index=False)
```
